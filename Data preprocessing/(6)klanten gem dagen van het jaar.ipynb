# Plotten: Totaal gemiddelde pakketten, per klant, per dag, op jaarbasis 

## 1.1 data manipuleren

```python
#import packages
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker

#data inladen
df = pd.read_csv('/datc/parcel/notebooks/data/postnl/20200904_sample.csv', index_col=0, parse_dates=[1])

#data manipuleren
df['dag'] = pd.PeriodIndex(df['procesdag'], freq='D').dayofyear
new_df = df.groupby([df['cust_id'], df['dag']]).agg(
    gem_aantal_pakketten = ('aantal_pakketten', 'mean'),
    gem_aantal_pakketten_volgende_dag = ('aantal_pakketten_volgende_dag', 'mean')).reset_index()
```

## 1.2 plotten

```python
#x en y, 1ste plot (klant 1)
df_klant1 = new_df.where(new_df['cust_id'] == 'klant_1').dropna()
fig, ax = plt.subplots()
ax.bar(df_klant1['dag'], df_klant1['gem_aantal_pakketten'])
ax.set(
    xlabel = 'Dag',
    ylabel = 'Aantal pakketten',
    title = 'Gemiddelde aantal pakketten per dag op jaarbasis voor klant 1')
plt.show()
```

## 1.3 conclusie

```python
#Het gemiddelde aantal pakketten ligt rond de 6000.
#We zien een redelijk regelmatige lijn. Alleen rond de 350ste dag is er een overduidelijke piek, 
#deze kan te verklaren zijn doordat het het weekend voor kerst(avond) is.
```
